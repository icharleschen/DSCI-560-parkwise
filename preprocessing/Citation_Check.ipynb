```python
import ast
import geopy.distance
import pandas as pd
```

# 6 Random Points in LA

```python
# Read 6 points
six_points = pd.read_csv('6 Random Points for Crime and Citation.txt', sep=', ', 
                        names=["lat", "lon", "location"], engine='python')
```

```python
display(six_points)
```

# Citation Data

```python
# Read preprocessed data
citation_df = pd.read_csv("citation_code_aggregation_2023.csv")
```

```python
display(citation_df)
```

# Calculate Distance between 6 Points and Entires in Crime Data

```python
# Convert entires from 6 points to list of coordinates
destinations = six_points[['lat', 'lon', 'location']].values.tolist()
print(destinations)
```

```python
for location in destinations:
    distance_list = list()
    for index, row in citation_df.iterrows():
        # Format start and end location
        start_location = ast.literal_eval(row['lat_lon'])
        start_coor = (float(start_location[0]), float(start_location[1]))
        end_coor = (location[0], location[1])
        
        # Calculate distance in miles
        distance = geopy.distance.geodesic(start_coor, end_coor).miles
        distance_list.append(round(distance, 3))

    # Add to crime df as new column
    citation_df['distance_to_' + str(location[2])] = distance_list
```

```python
# citation_df
```

```python
df_WOF = citation_df[citation_df['distance_to_Walk Of Fame'] <= 2.0].iloc[:, :-6]
df_USC = citation_df[citation_df['distance_to_USC'] <= 2.0].iloc[:, :-6]
df_CC = citation_df[citation_df['distance_to_Culver City'] <= 2.0].iloc[:, :-6]
df_LB = citation_df[citation_df['distance_to_Long Beach'] <= 8.0].iloc[:, :-6]  # Changed filter distance
df_SM = citation_df[citation_df['distance_to_Santa Monica'] <= 2.0].iloc[:, :-6]
df_DTLA = citation_df[citation_df['distance_to_DTLA'] <= 2.0].iloc[:, :-6]
```

```python
df_WOF['All Violation'] = df_WOF.iloc[:, 2:].sum(axis=1)
# df_WOF.head()
df_USC['All Violation'] = df_USC.iloc[:, 2:].sum(axis=1)
df_CC['All Violation'] = df_CC.iloc[:, 2:].sum(axis=1)
df_LB['All Violation'] = df_LB.iloc[:, 2:].sum(axis=1)
df_SM['All Violation'] = df_SM.iloc[:, 2:].sum(axis=1)
df_DTLA['All Violation'] = df_DTLA.iloc[:, 2:].sum(axis=1)
```

```python
df_WOF_agg = df_WOF.groupby(['week']).sum().reset_index()
df_WOF_agg['location'] = 'Walk Of Fame'

df_USC_agg = df_USC.groupby(['week']).sum().reset_index()
df_USC_agg['location'] = 'USC'

df_CC_agg = df_CC.groupby(['week']).sum().reset_index()
df_CC_agg['location'] = 'Culver City'

df_LB_agg = df_LB.groupby(['week']).sum().reset_index()
df_LB_agg['location'] = 'Long Beach'

df_SM_agg = df_SM.groupby(['week']).sum().reset_index()
df_SM_agg['location'] = 'Santa Monica'

df_DTLA_agg = df_DTLA.groupby(['week']).sum().reset_index()
df_DTLA_agg['location'] = 'DTLA'
```

```python
# df_DTLA_agg
```

```python
df_final = pd.concat([df_WOF_agg, df_USC_agg, df_CC_agg, df_LB_agg, df_SM_agg, df_DTLA_agg])
df_final = df_final.reset_index(drop=True)
# display(df_final)
```

```python
# Save result
df_final.to_csv('citation_check_updated_2023.csv', index=False, encoding='utf-8')
```
