```python
import pandas as pd
```

```python
# Read preprocessed data
df = pd.read_csv("citation_data_complete.csv")
```

# Analysis on Violation Code

```python
unique_violation = df['violation_code'].unique()
print("Number of unique violation code: ", len(unique_violation))
```

```python
print(unique_violation)
```

```python
# Create a dataframe with only valid violation description
work_df = df[df['violation_description'].notna()]
```

```python
%%time
# Obtain dictionary for each unique violation code
violation_dict = dict()
for code in unique_violation:
    try:
        description = work_df.loc[work_df['violation_code'] == code, 'violation_description'].iloc[0]
        violation_dict[code] = description
    except:
        violation_dict[code] = ''
```

```python
# Preview of dictionary
# {k: violation_dict[k] for k in list(violation_dict)}
```

```python
# Keep key-value pair with violation description
valid_violation_dict = {k: v for k, v in violation_dict.items() if v}
{k: valid_violation_dict[k] for k in list(valid_violation_dict)}
```

```python
%%time
# Convert issue date columns to datetime
"""
This step is done in the preprocessing, but when the issue date column is read as object data type when load from local
"""
df['issue_date'] = df['issue_date'].apply(pd.to_datetime)
```

```python
# Create lat_lon column that contains latitude AND longitude information
df['lat_lon'] = df.apply(lambda x: (x.lat, x.lon), axis=1)
# display(df)
```

```python
df.dtypes
```

# Aggregation on Location and Different Violation Activities

```python
# Create new column and round lat_lon to 2 decimal place
df['lat_lon_general'] = df.apply(lambda x: ("%.2f" % round(x.lat, 2), "%.2f" % round(x.lon, 2)), axis=1)
```

```python
# Drop rows with Nan in violation_description column
df = df[df['violation_description'].notna()]
display(df)
```

```python
df_agg = df.groupby([df.issue_date.dt.isocalendar().year, 
                     df.issue_date.dt.isocalendar().week, 
                     df.lat_lon_general, 
                     df.violation_code]).agg({'violation_code': 'count'})
df_agg.columns = ['violation_count']
# display(df_agg)
```

```python
# Rename axis and flatten the result
df_agg_flat = df_agg.rename_axis(['year', 'week', 
                                  'lat_lon', 'violation_code']).reset_index()

# display(df_agg_flat)
```

```python
# Pivot table
df_violation_code = df_agg_flat.pivot(index=['year', 'week', 'lat_lon'], 
                                      columns=['violation_code'], 
                                      values='violation_count')
# Reset index and replace NaN
df_violation_code_flat = df_violation_code.reset_index()
df_violation_code_flat.fillna(0, inplace=True)
# display(df_violation_code_flat)
```

```python
# Filter on year and week
df_violation_code_flat = df_violation_code_flat[df_violation_code_flat['year']==2023]
# df_violation_code_flat = df_violation_code_flat[df_violation_code_flat['week'] >= 8]
```

```python
# Create column dict to format aggregation
column_name = list(df_violation_code_flat.columns[3:])
agg_dict = {x:'sum' for x in column_name}
# print(agg_dict)
```

```python
df_violation_code_agg = df_violation_code_flat.groupby(['week', 'lat_lon']).agg(agg_dict)
# display(df_violation_code_agg)
```

```python
# Reset axis
df_violation_code_agg = df_violation_code_agg.rename_axis(['week', 'lat_lon']).reset_index()
# Rename columns
valid_violation_dict['014'] = 'PARKED IN BUS LOADING ZONE'
valid_violation_dict['000'] = 'PARKING PROHIBITED OR LIMITED'
df_violation_code_agg.rename(columns=valid_violation_dict, inplace=True)
# display(df_violation_code_agg)
```

```python
# Save to csv
df_violation_code_agg.to_csv('citation_code_aggregation_2023.csv', index=False, encoding='utf-8')
```
