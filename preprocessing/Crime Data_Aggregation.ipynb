```python
import numpy as np
import pandas as pd
```

```python
# Read preprocessed data
df = pd.read_csv("crime_data_complete.csv")
```

```python
display(df)
```

# Analysis on Crime Code

```python
unique_crimecd = df['crm_cd'].unique()
print("Number of unique crime code: ", len(unique_crimecd))
```

```python
print(unique_crimecd)
```

```python
%%time
# Obtain dictionary for each unique crime code
crimecode_dict = dict()
for code in unique_crimecd:
    try:
        description = df.loc[df['crm_cd'] == code, 'crm_cd_desc'].iloc[0]
        crimecode_dict[code] = description
    except:
        crimecode_dict[code] = ''
```

```python
# Preview crime code and description
crimecode_dict
```

```python
# Convert columns to datetime
df[['date_rptd', 'date_occ']] = df[['date_rptd', 'date_occ']].apply(pd.to_datetime)
```

```python
# Create lat_lon column that contains latitude AND longitude information
df['lat_lon'] = df.apply(lambda x: (x.lat, x.lon), axis=1)
```

```python
display(df)
```

```python
df.dtypes
```

# Aggregation on Location and Different Criminal Activities

```python
# Create new column and round lat_lon to 2 decimal place
df['lat_lon_general'] = df.apply(lambda x: ("%.2f" % round(x.lat, 2), "%.2f" % round(x.lon, 2)), axis=1)
```

```python
display(df)
```

```python
df_agg = df.groupby([df.date_occ.dt.isocalendar().year, 
                            df.date_occ.dt.isocalendar().week, 
                            df.lat_lon_general, 
                            df.crm_cd]).agg({'crm_cd': 'count'})
```

```python
display(df_agg)
```

```python
df_agg.columns = ['crime_count']
# display(df_agg)
```

```python
df_agg_flat= df_agg.rename_axis(['year', 'week', 
                                 'lat_lon', 'crime_code']).reset_index().rename(columns={'crm_cd':'crime_count'})
# display(df_agg_flat)
```

```python
df_crime_code = df_agg_flat.pivot(index=["year", "week", "lat_lon"], columns=['crime_code'], values='crime_count')
# display(df_crime_code)
```

```python
df_crime_code_flat = df_crime_code.reset_index()
# display(df_crime_code_flat)
```

```python
# Replace NaN with 0
df_crime_code_flat.fillna(0,inplace=True)

# Filter on year
df_crime_code_flat = df_crime_code_flat[df_crime_code_flat['year']==2023]
# df_crime_code_flat = df_crime_code_flat[df_crime_code_flat['week'] >= 8]

# display(df_crime_code_flat)
```

```python
column_name = list(df_crime_code_flat.columns[3:])
# print(column_name)
agg_dict = {x:'sum' for x in column_name}
# print(agg_dict)
```

```python
df_crime_code_agg = df_crime_code_flat.groupby(['week', 'lat_lon']).agg(agg_dict)
# display(df_crime_code_agg)
```

```python
df_crime_code_agg = df_crime_code_agg.rename_axis(['week', 'lat_lon']).reset_index()
```

```python
df_crime_code_agg.rename(columns = crimecode_dict, inplace=True)
```

```python
# Save to csv
df_crime_code_agg.to_csv('crime_code_aggregation_2023.csv', index=False, encoding='utf-8')
```
